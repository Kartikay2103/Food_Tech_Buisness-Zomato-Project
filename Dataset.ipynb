```python
# Checking Info of Data and their Datatypes
df.info()
```

```python
# Converting Order_Date Datatype to Datetime
df['Order_Date '] = pd.to_datetime(df['Order_Date'], format = '%d-%m-%Y')
```

```python
# Dropping Unnecessary columns
drop_cols =['Restaurant_latitude','Restaurant_longitude','Delivery_location_latitude','Delivery_location_longitude','Delivery_person_ID']
df.drop(drop_cols,axis = 1,inplace = True)
```

```python
# Checking Shape
df.shape
```

```python
# Checking null values
df.isnull().sum()
```

```python
# Filling null values with mode
df.fillna({
    'Delivery_person_Age': df['Delivery_person_Age'].mode()[0],
    'Delivery_person_Ratings': df['Delivery_person_Ratings'].mode()[0],
    'Time_Orderd': df['Time_Orderd'].mode()[0],
    'Weather_conditions': df['Weather_conditions'].mode()[0],             
    'Road_traffic_density': df['Road_traffic_density'].mode()[0],
    'multiple_deliveries': df['multiple_deliveries'].mode()[0],
    'Festival': df['Festival'].mode()[0],
    'City': df['City'].mode()[0]
},inplace = True)
```

```python
# There is no duplicated data in data
df.duplicated().sum()
```

```python
# Dropping cloumn of Delivery_person_age
df.drop(['Delivery_person_Age'], axis = 1, inplace = True)
```

```python
# Convert categorical columns to category dtype (reduces size)
categorical_cols = ['Weather_conditions','Road_traffic_density','Type_of_order','Type_of_vehicle','Festival','City']
df[categorical_cols] = df[categorical_cols].astype('category')
```

```python
# Convert Order_Date to datetime (date format)
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y', errors='coerce')

# Convert Time columns to datetime (time format only)
df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'], format='%H:%M', errors='coerce')
df['Time_Order_picked'] = pd.to_datetime(df['Time_Order_picked'], format='%H:%M', errors='coerce')
```

```python
df.info()
```

```python
df.to_excel("final_clean_dataset.xlsx", index=False, engine='openpyxl')
```
